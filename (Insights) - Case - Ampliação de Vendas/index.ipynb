## **Python (Insights) - Análise de Dados com Python**

### **Case: Ampliação de Vendas**

**Desafio**

*Você trabalha em uma empresa que vende bermudas e tem cinco lojas na cidade. O diretor percebeu que as vendas estão abaixo do esperado e quer aumentar o faturamento, especialmente com o verão chegando. O que você faria para atrair mais clientes e impulsionar as vendas?*


Base de dados: [Vendas xlsx](./Vendas.xlsx)

Referência: [youtube](https://www.youtube.com/watch?v=kCMaqla6Grs&list=PLpdAy0tYrnKx9CtTmgSdzHz9YQ-C5ZNI9)

```python
# Passo a passo

# (1) - Importar Base de dados.
# (2) - Visualizar e Tratar a Base de dados.
# (3) - Observar base p/ obter dados que resovam o desafio.
# (4) - Construir análise p/ chegar a uma decisão que impulsionaria as vendas.
```

```python
# Passo - 1 (Importar Base de dados):

import pandas as pd

base = pd.read_excel('./Vendas.xlsx')
display(base)
```

```python
# Passo - 2 (Visualizar e Tratar a Base de dados):

faturamento = base['Valor Final'].sum()
print(faturamento)
```

```python
# Passo - 3 (Observar base p/ obter dados que resovam o desafio):

# Faturamento por Loja
print('\nFaturamento por Loja')
faturamento_por_loja = base[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
display(faturamento_por_loja)

# Faturamento por Produto
print('\nFaturamento por Loja')
faturamento_por_produto = base[['ID Loja', 'Produto', 'Valor Final']].groupby(['ID Loja', 'Produto']).sum()
display(faturamento_por_produto)
```

```python
# Passo - 4 (Construir análise p/ identificar principais motivos de Cancelamento.)
```

# Conclusão

   - O faturamento das vendas foi maior nas lojas que tinham o produto 'Bermuda Liso'.
    ##### Recomenda-se adicionar o produto 'Bermuda Liso' ao estoque disponível para vendas em todas as lojas.
